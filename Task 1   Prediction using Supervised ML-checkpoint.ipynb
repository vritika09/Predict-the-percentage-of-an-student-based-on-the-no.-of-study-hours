```python
#import the requred libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
```

```python
#Read the dataset
data = pd.read_csv("http://bit.ly/w-data")

print("Data Imported Sucessfully !!")

print(data.head())
```

```python
#Check if there any null value or not in our datset
data.isnull == True
```

so there is not null value in dataset. if isnull return True means null value present otherwise not

# Visualize Our data Using Matplotlib Library

```python
plt.scatter(x = data['Hours'], y = data['Scores'])
plt.grid(color='red')
plt.title("Marks Vs Study hours",size = 20)
plt.xlabel("Studies(Hours)",size=12)
plt.ylabel("Marks(Percentage)",size=12)
plt.show()
```

Form the above scatter plot there looks to be correlations between the ' Marks' and 'Hours Study'. Lets plot a regression line to confirm the correlation.

```python
sns.regplot(x = data['Hours'], y = data['Scores'])
plt.title("Regression Plot", size = 20)
plt.ylabel("Marks(Precentage)",size = 13)
plt.xlabel("Hours(study)", size=13)

plt.show()

print(data.corr())
```

So Through this Regression plot we have to confirmed that the variables are positively correlated.

# Train The Model

```python
#divide the data into x and y 
X = data.iloc[:, :-1].values
Y = data.iloc[:, 1].values
```

```python
#spliting the data into training and testing
train_X, test_X, train_Y, test_Y = train_test_split(X,Y,random_state=0)
```

# Fitting the Data into the Model

```python
regression = LinearRegression()
regression.fit(train_X,train_Y)

print(' Model Trained '.center(50,'*'))
```

# Predicting the Percentage of Marks¶

```python
pred_y = regression.predict(test_X)
prediction = pd.DataFrame({'Hours': [i[0] for i in test_X], 'Predicted Marks': [k for k in pred_y]})
prediction
```

# Comparing the Predicted Marks with the Actual Marks

```python

compare_scores = pd.DataFrame({'Actual Marks': test_Y, 'Predicted Marks': pred_y})
compare_scores
```

# Visually Comparing the Predicted Marks with the Actual Marks

```python
plt.scatter(x=test_X, y=test_Y, color='red')
plt.plot(test_X, pred_y, color='Black')
plt.title('Actual vs Predicted', size=20)
plt.ylabel('Marks (Percentage)', size=12)
plt.xlabel('Hours (Studied)', size=12)
plt.show()
```

# Evaluating the Model

```python
# Calculating the accuracy of the model
print('Mean absolute error: ',mean_absolute_error(test_Y,pred_y))
```

# What will be the predicted score of a student if he/she studies for 9.25 hrs/ day?

```python
hours = [9.25]
answer = regression.predict([hours])
print("Score = {}".format(round(answer[0],3)))
```

According to the regression model if a student studies for 9.25 hours a day he/she is likely to score 93.89 marks.
